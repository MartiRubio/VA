### 1.1 Image creation, saving and loading

Remember that [Numpy](https://docs.scipy.org/doc/numpy-1.13.0/reference/) allows us to define images of different kinds treating them as matrices.

The most common examples are:

* Binary: with values of 0 or 1, interpreted as black or white, respectively.
* Grayscale: with possible data types uint8, uint16, double, whose pixel values define their intensity.
* RGB Image: color images consisting of 3 matrices (usually called 'channels'), which contain intensity values for each color separately R (red), G (green) and B (blue).


a) 
- Create and display a grayscale image that has 150 rows and 200 columns where each value should be of type unsigned integer with 8 bits. 

```python
# A matrix is created with random numbers between 0 and 1
grayscale = np.random.rand(150, 200)

# The range is rescaled to (0, 255)
grayscale *= 255

# The float is changed to uint8 data type 
grayscale = grayscale.astype('uint8')

# The image is shown
plt.imshow(grayscale, cmap='gray')
plt.show()
```

```python
# The darker values of the gray_Lenna image are selected
lenna_dark = lenna_gray[:,:] < 100./255.

# The brighter values of the gray_Lenna image are selected
lenna_bright = lenna_gray[:,:] > 100./255.

# The darker values are blackened and saved in lenna_baw image
lenna_baw = lenna_gray*0

# The brighter values are whitened and saved in lenna_baw image
lenna_baw[lenna_bright] += 255.

# The image is shown
plt.imshow(lenna_baw, cmap='gray')
plt.show()

# The different colour values in the image are displayed (black and white)
print('Unique colours: ', np.unique(lenna_baw))

# The tenth row is printed
print('\nTenth row: \n', lenna_baw[10])
```
